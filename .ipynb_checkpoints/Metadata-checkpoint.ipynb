```python
from pandas import DataFrame
import pandas as pd
```

# Metadata Jagung

```python
#readcsv
df = pd.read_csv('Sample_Jagung.csv', index_col=0)
df.head()
```

```python
#pick random sample, if sample < 3, pick all
df_out = pd.DataFrame()
for i in df.ID_Study.unique():
    #print(i)
    data = df[df.ID_Study == i]
    if len(data) < 3:
        data = data
    else:
        data = data.sample(n=3)
    df_out = df_out.append(data)
df_out
```

```python
#create metadata container
df_meta = DataFrame(columns=('#SampleID',
                             'BarcodeSequence',
                             'LinkerPrimerSequence', 
                             'StudyID',
                             'SampleID',
                             'SampleName',
                             'Latitude',
                             'Longitude',
                             'Geography',
                             'Biome', 
                             'Depth',
                             'Elevation',
                             'pH',
                             'Description'))
df_meta
```

```python
#fill in metadata
df = df_out
for i in df.index:
    df_meta.loc[i]  = [df.loc[i, 'ID_Analyses'],\
                  '_',\
                  '_',\
                  df.loc[i, 'ID_Study'],\
                  df.loc[i, 'ID_Sample'],\
                  df.loc[i, 'Sample Name'],\
                  df.loc[i, 'Latitude'],\
                  df.loc[i, 'Longitude'],\
                  df.loc[i, 'geographic location'],\
                  df.loc[i, 'environment (biome)'],\
                  df.loc[i, 'depth (m)'],\
                  df.loc[i, 'elevation (masl)'],\
                  df.loc[i, 'pH'],\
                  'Corn']
df_meta
```

```python
#export metadata
df_meta.to_csv('metadata_jagung.txt', sep="\t", index = False)
```
